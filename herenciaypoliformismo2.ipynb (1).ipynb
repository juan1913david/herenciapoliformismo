```python
class Animal():
    def __init__(self):
        self.__Color=""
        self.__Peso=""
        self.__Comer=""
        self.__Moverse=""

    def GetColor(self):
      return self.__Color
  
    def SetColor(self, color):
     self.__Color=color
  
    def GetPeso(self):
     return self.__Peso

    def SetPeso(self, peso):
     self.__Peso = peso

    def GetComer(self):
     return self.__Comer

    def SetComer(self, comer):
     self.__Comer=comer

    def GetMoverse(self):
      return self.__Moverse

    def SetMoverse(self, moverse):
     self.__Moverse = moverse

class Ave(Animal):
  def __init__(self):
    self.__Tipo=""
    self.__Volar=""

  def GetTipo(self):
    return self.__Tipo

  def SetTipo(self, tipo):
    self.__Tipo = tipo

  def GetVolar(self):
    return self.__Volar

  def SetVolar(self, volar):
    self.__Volar = volar

  def Mostrardatos_ave(self):
    print("(AVE):")
    print("\tTipo de ave: ",self.__Tipo)
    print("\tColor del ave: ",self.GetColor())
    print("\tPeso del ave: ",self.GetPeso())
    print("\tpuede volar a traves de sus: ",self.__Volar)
    print("\tpuede desplazarce a traves de sus: ",self.GetMoverse())
    print("\tpuede alimentarse a traves de su: ",self.GetComer())
    
ave = Ave()
ave.SetTipo("Serinus canaria")
ave.SetColor("Multicolores")
ave.SetPeso("3kg")
ave.SetVolar("Alas-Plumas")
ave.SetMoverse("2 Patas")
ave.SetComer("Pico-boca")
ave.Mostrardatos_ave()


class Perro(Animal):
  def __init__(self):
    self.__Raza=""
    self.__Ladrar=""

  def GetRaza(self):
    return self.__Raza

  def SetRaza(self, raza):
    self.__Raza = raza

  def GetLadrar(self):
    return self.__Ladrar

  def SetLadrar(self, ladrar):
    self.__Ladrar = ladrar

  def Mostrardatos_perro(self):
    print("(PERRO):")
    print("\tRaza del perro: ",self.__Raza)
    print("\tColor del perro: ",self.GetColor())
    print("\tPeso del perro: ",self.GetPeso())
    print("\tladra gracias a sus : ",self.__Ladrar)
    print("\tpuede desplazarse con sus: ",self.GetMoverse())
    print("\tpuede alimentarce atraves de su: ",self.GetComer())
    
perro = Perro()
perro.SetRaza("Pitbull")
perro.SetColor("Blanco")
perro.SetPeso("20kg")
perro.SetLadrar("Cuerdas Vocales")
perro.SetMoverse("4 Patas")
perro.SetComer("Boca(hocico)")
perro.Mostrardatos_perro()
```

```python
#se puede hacer polimorfismo con metodo for 
class Colombia:
  def capital(self):
    print("bogota")
  def idioma(self):
    print("español")
class Francia:
  def capital(self):
    print("paris")
  def idioma(self):
    print("frances")

colombiano = Colombia() 
frances = Francia()
for pais in (colombiano,frances):
  pais.capital()
  pais.idioma()
```

```python
#y polimorfismo con función
class Tomate:
  def tipo(Self):
    print("vegetal")
  def color(self):
    print("rojo")

class Manzana:
  def tipo(self):
    print("fruta")
  def color(self):
    print("verde")

def funcion(objeto):
  objeto.tipo()
  objeto.color()
nuevo_tomate = Tomate()
funcion(nuevo_tomate)
nueva_manzana = Manzana()
funcion(nueva_manzana)
```
